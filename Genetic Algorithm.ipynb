# Implementation of Genetic Algorithm

```python
import numpy as np
```

### Generate Chromosome

```python
def chromosome(gene_size = 20):
    c = np.random.randint(2, size= gene_size)
    return c
```

### Generate Population

```python
populate = []
for i in range(30):
    new_c = chromosome(20)
    populate.append(new_c)
    
population = np.array(populate)
```

### Fintess

```python
def fitness(a):
    x = a.dot(1 << np.arange(a.shape[-1] - 1, -1, -1))
    return x
```

### Rank Selection

```python
def rank_selection(population):
    fit_value = {}
    for i in range(30):
        fit_value[i] = fitness(population[i])
    fit_value = {k: v for k, v in sorted(fit_value.items(), key=lambda item: item[1])}
    return fit_value 
```

### CrossOver

```python
def cross_over(parent_1, parent_2):
    while True:
        i = np.random.choice(20)
        j = np.random.choice(20)
        if j != i:
            break
    
    if i>j:
        temp = i
        i = j
        j = temp
    cross_segment_1 = parent_1[i:j].copy()
    cross_segment_2 = parent_2[i:j].copy()
    
    parent_1[i:j] = cross_segment_2.copy()
    parent_2[i:j] = cross_segment_1.copy()
    
    return parent_1,parent_2
        
```

### Mutation

```python
def mutation(offspring):
    i = np.random.choice(20)
    if offspring[i] == 0:
        offspring[i] = 1
    else:
        offspring[i] = 0
    
    return offspring
```

### Elemenation: Survival of the fittest 

```python
def elemenation(population,fit_value,offspring):
    if (list(fit_value.values())[29])>fitness(offspring):
        population[29] = offspring.copy()
    
    return population
        
```

## Genetic Algorithm

```python
def gen_algo(generations,population):
    
    for gen in range(generations):
        
        print("Generation : ",gen,"\n\n")
        
        #selection of the fittest parents
        fit_value =  rank_selection(population)
    
        #two parents
        parent_1 = population[list(fit_value.keys())[0]].copy()
        parent_2 = population[list(fit_value.keys())[1]].copy()
        
        print("Selected Parent-1: ",parent_1," = ",fitness(parent_1),"\n")
        print("Selected Parent-2: ",parent_2," = ",fitness(parent_2),"\n\n")
    
        #crossover and new children
        new_child_1, new_child_2 = cross_over(parent_1,parent_2)
        
        print("Cross over Child_1: ",new_child_1," = ",fitness(new_child_1),"\n")
        print("Cross over Child_2: ",new_child_2," = ",fitness(new_child_2),"\n\n")
    
    
        #mutation of any random child
        t = np.random.randint(11)
        r = np.random.choice(2)
        if(t%9==0):
            if(r == 0):
                final_new_child = mutation(new_child_1)
        
            else:
                final_new_child = mutation(new_child_2)
        
            print("After Mutation, Final Child: ",final_new_child," = ",fitness(final_new_child),"\n")
        
        else:
            if(r == 0):
                final_new_child = new_child_1.copy()
        
            else:
                final_new_child = new_child_2.copy()
            
            print("No Mutation")
            
        population = elemenation(population,fit_value,final_new_child)
        print("New Entry: ",population[29]," = ",fitness(population[29]),"\n")
        
        print("------------------------------------------------------------------------------------------------------\n\n")
        
        
    print("-------------------------------------------Execution Finished-------------------------------------------------")    
    return population
    
```

```python
result_population = gen_algo(100, population)
```
